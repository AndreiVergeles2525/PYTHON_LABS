```python
from math import isqrt

def is_int_token(t: str) -> bool:
    try:
        int(t)
        return True
    except ValueError:
        return False

def is_prime(n: int) -> bool:
    if n < 2:
        return False
    if n in (2, 3):
        return True
    if n % 2 == 0:
        return False
    r = isqrt(n)
    i = 3
    while i <= r:
        if n % i == 0:
            return False
        i += 2
    return True

def task_6():
    line = input().strip()
    parts = line.split()
    if len(parts) != 2:
        print("Помилка: потрібно ввести рівно два значення через пробіл")
        return None
    if not all(is_int_token(p) for p in parts):
        print("Помилка: значення мають бути цілими числами")
        return None
    a, b = map(int, parts)
    if a > b:
        print("Помилка: a має бути меншим або рівним b")
        return None

    res = [n for n in range(a, b + 1) if is_prime(n)]
    return res
task_6()
task_6()
task_6()
```
